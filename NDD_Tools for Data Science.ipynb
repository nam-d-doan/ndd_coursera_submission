# Arithmetic Expressions Examples

Arithmetic expressions are mathematical statements that perform basic arithmetic operations. These include addition, subtraction, multiplication, division, and exponentiation. Below are some examples of arithmetic expressions:

1. **Addition (+)**: This operation adds two numbers together.
   - Example: `3 + 5 = 8`

2. **Subtraction (-)**: This operation subtracts one number from another.
   - Example: `10 - 4 = 6`

3. **Multiplication (*)**: This operation multiplies two numbers.
   - Example: `7 * 3 = 21`

4. **Division (/)**: This operation divides one number by another.
   - Example: `20 / 4 = 5`

5. **Exponentiation (^ or **)**: This operation raises one number to the power of another.
   - Example: `2^3 = 8` or `2**3 = 8`

## Combined Operations

Arithmetic expressions can combine multiple operations, and the order of operations (PEMDAS/BODMAS) is important to get the correct result:
- **Parentheses/Brackets**
- **Exponents/Orders**
- **Multiplication and Division**
- **Addition and Subtraction**

Examples:
- `3 + 2 * 5 = 13` (Multiplication before addition)
- `(3 + 2) * 5 = 25` (Parentheses first)
- `4^2 / 2 = 8` (Exponentiation before division)
